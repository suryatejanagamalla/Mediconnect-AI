```python
import os
import whisper
from whisper.utils import get_writer
```

```python
model = whisper.load_model("base")
```

```python
input_directory = "./audio_files"
output_directory = "./transcriptions"
os.makedirs(output_directory, exist_ok=True)
```

```python
for filename in os.listdir(input_directory):
    if filename.endswith(".mp3") or filename.endswith(".wav"):
        audio_path = os.path.join(input_directory, filename)
        print(f"Processing {filename}")
```

```python
result = model.transcribe(audio_path)
```

```python
transcription_no_breaks = os.path.join(output_directory, f"{os.path.splitext(filename)[0]}_no_breaks.txt")
with open(transcription_no_breaks, "w", encoding="utf-8") as txt:
            txt.write(result["text"])
```

```python
txt_writer = get_writer("txt", output_directory)
txt_writer(result, audio_path)
```
